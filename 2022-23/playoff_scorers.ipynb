```python
regular_all_players = leaguedashplayerstats.LeagueDashPlayerStats(per_mode_detailed='PerGame', season_type_all_star='Regular Season').get_data_frames()[0]

time.sleep(0.5)

advanced_regular_all_players = leaguedashplayerstats.LeagueDashPlayerStats(measure_type_detailed_defense='Advanced', per_mode_detailed='PerGame', season_type_all_star='Regular Season').get_data_frames()[0]

regular_all_players = pd.merge(regular_all_players, advanced_regular_all_players, on=["PLAYER_ID", "TEAM_ID", 'PLAYER_NAME'])
```

```python
playoff_all_players = leaguedashplayerstats.LeagueDashPlayerStats(per_mode_detailed='PerGame', season_type_all_star='Playoffs').get_data_frames()[0]

time.sleep(0.5)

advanced_playoff_all_players = leaguedashplayerstats.LeagueDashPlayerStats(measure_type_detailed_defense='Advanced', per_mode_detailed='PerGame', season_type_all_star='Playoffs').get_data_frames()[0]

playoff_all_players = pd.merge(playoff_all_players, advanced_playoff_all_players, on=["PLAYER_ID", "TEAM_ID", 'PLAYER_NAME'])
```

```python
playoff_all_players = playoff_all_players.sort_values(by=['PTS'], ascending=False)

top10_playoff_scorers = playoff_all_players[:10]

westbrook_id = 201566

top10_playoff_scorers = pd.concat([top10_playoff_scorers, playoff_all_players.loc[playoff_all_players['PLAYER_ID'] == westbrook_id]])

top10_playoff_scorers
```

```python
def drawArrow(A, B):
    plt.arrow(A[0], A[1], B[0] - A[0], B[1] - A[1],
              head_width=0.01, length_includes_head=False)
```

```python
zoom = 0.275
player_root_dir = 'player_imgs/'

def get_player_img(player_id):
    #print(path)
    return OffsetImage(plt.imread(f'{player_root_dir}{player_id}.png'), zoom=zoom)

fig, ax = plt.subplots(figsize=(18, 18))


top10_regular_scorers_ids = top10_regular_scorers['PLAYER_ID']

top10_playoff_scorers_ids = top10_playoff_scorers['PLAYER_ID']


playoff_scorers_from_top10_regular_scorers = pd.DataFrame()

for id_ in top10_regular_scorers_ids:
    playoff_scorers_from_top10_regular_scorers
    playoff_scorers_from_top10_regular_scorers = pd.concat([playoff_scorers_from_top10_regular_scorers, playoff_all_players.loc[playoff_all_players['PLAYER_ID'] == id_]])

for id_ in top10_playoff_scorers_ids:
    playoff_scorers_from_top10_regular_scorers = pd.concat([playoff_scorers_from_top10_regular_scorers, playoff_all_players.loc[playoff_all_players['PLAYER_ID'] == id_]])

# ax.scatter(top10_regular_scorers['FG_PCT'], y=top10_regular_scorers['PTS'], s=10, label='Regular Season')
# ax.scatter(playoff_scorers_from_top10_regular_scorers['FG_PCT'], playoff_scorers_from_top10_regular_scorers['PTS'], s=10, label='Playoff')

# plt.show()

get_value = lambda x: x.EFG_PCT

for i in range(len(playoff_scorers_from_top10_regular_scorers)):
    playoff_item = playoff_scorers_from_top10_regular_scorers.iloc[i]
    player_id = playoff_item.PLAYER_ID
    player_name = playoff_item.PLAYER_NAME
    
    if player_id == 1628369 or player_id == 1629027:
        continue
    
    regular_item = regular_all_players.loc[regular_all_players['PLAYER_ID'] == player_id].squeeze()
    
    ax.scatter(get_value(regular_item) * 100, regular_item.PTS, color='r', s=10, label='Regular Season')
    ax.scatter(get_value(playoff_item) * 100, playoff_item.PTS, color='b', s=10, label='Playoff')


    A = (get_value(regular_item) * 100, regular_item.PTS)
    B = (get_value(playoff_item) * 100, playoff_item.PTS)
    
    if B[1] > A[1]:
        color = 'g'
    else:
        color = 'r'

    style = ArrowStyle('Fancy', head_length=7, head_width=5)
    arrow = FancyArrowPatch(A, B, mutation_scale=2, arrowstyle=style, color=color, shrinkB=20, linewidth=1.5)                     
    ax.add_patch(arrow)
    
    # regular_offsetbox = TextArea(f"{player_name}")
    # playoff_offsetbox = TextArea(f"{player_name}")
    # ab_regular = AnnotationBbox(playoff_offsetbox, (regular_item.FG_PCT, regular_item.PTS), frameon=False)
    # ab_playoff = AnnotationBbox(regular_offsetbox, (playoff_item.FG_PCT, playoff_item.PTS), frameon=False)

    ab_regular = AnnotationBbox(get_player_img(player_id), (get_value(playoff_item) * 100, playoff_item.PTS), frameon=False)
    ab_playoff = AnnotationBbox(get_player_img(player_id), (get_value(regular_item) * 100, regular_item.PTS), frameon=False)
    ax.add_artist(ab_regular)
    ax.add_artist(ab_playoff)
    
    diff = np.round(playoff_item.PTS - regular_item.PTS, 2)
    if diff > 0:
        print(f'{player_name} : +{diff}')
    else:
        print(f'{player_name} : {diff}')

# plt.rcParams.update({'font.size': 25})
    
ax.set_xlabel("EFG", fontsize=25)
ax.set_ylabel("PPG", fontsize=25)
# ax.set_xticks()
# ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.xaxis.set_major_formatter(mtick.PercentFormatter())
ax.set_title("Better or worse playoff scorer?", fontdict={'fontsize': 35})
ab = AnnotationBbox(OffsetImage(plt.imread(f'twitter_logo (1).png'), zoom=0.1), (65.2, 34.8), frameon=False)
ax.add_artist(ab)
ax.text(x=67, y=34.77, ha='right', s='@FHuzjan', size=15)

# plt.savefig('better_worse_playoff_scorer.png', bbox_inches='tight')
```
