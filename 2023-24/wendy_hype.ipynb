```python
def convert_season_to_int(season):
    season = season.split('-')[0]
    return int(season) + 1

def convert_int_to_season(year):
    previos_year = year - 1
    year = str(year)
    previous_year = str(previos_year)
    year = year[-2:]
    return f'{previos_year}-{year}'
```

```python
zoom = 0.275
player_root_dir = 'player_imgs/'

def get_player_img(player_id):
    #print(path)
    return OffsetImage(plt.imread(f'{player_root_dir}{player_id}.png'), zoom=zoom)

fig, ax = plt.subplots(figsize=(18, 18))

seasons = list(map(convert_season_to_int, spurs['YEAR']))

ax.plot(seasons, spurs_win_pct * 100)

# ax.plot(x=list(map(convert_season_to_int, tim_duncan_carrer['SEASON_ID'])), y=)

timmy_drafted = (convert_season_to_int(tim_duncan_first_year), spurs[spurs['YEAR'] == tim_duncan_first_year]['WIN_PCT'] * 100)
timmy_ended = (convert_season_to_int(tim_duncan_last_year), spurs[spurs['YEAR'] == tim_duncan_last_year]['WIN_PCT'] * 100)

david_started = (convert_season_to_int(david_robinson_first_year), spurs[spurs['YEAR'] == david_robinson_first_year]['WIN_PCT'] * 100)
david_ended = (convert_season_to_int(david_robinson_last_year), spurs[spurs['YEAR'] == david_robinson_last_year]['WIN_PCT'] * 100)

dx = 1.2
dy = 0.02 * 100

for i, pair in enumerate([timmy_drafted, timmy_ended]):
    ab = AnnotationBbox(get_player_img(tim_duncan_id), pair, frameon=False)

    ax.add_artist(ab)
    
    # ab = AnnotationBbox(TextArea('Timmy'), (pair[0] + 10, pair[1] - 2), frameon=False)
    if i == 0:
        text = 'Drafted'
    else:
        text = 'Retired'
    offsetbox = TextArea(f"{text}")

    ab = AnnotationBbox(offsetbox, (pair[0] - dx, pair[1] + dy),
                    box_alignment=(0., 0.5),
                    arrowprops=dict(arrowstyle="->"), frameon=False)

    ax.add_artist(ab)
    
for i, pair in enumerate([david_started, david_ended]):
    ab = AnnotationBbox(get_player_img(david_robinson_id), pair, frameon=False)

    ax.add_artist(ab)
    
    if i == 0:
        text = 'Drafted'
    else:
        text = 'Retired'
    offsetbox = TextArea(f"{text}")

    ab = AnnotationBbox(offsetbox, (pair[0] - dx, pair[1] + dy),
                    box_alignment=(0., 0.5),
                    arrowprops=dict(arrowstyle="->"), frameon=False)
    
    ax.add_artist(ab)

    

wendy_drafted = (convert_season_to_int('2022-23'), spurs[spurs['YEAR'] == '2022-23']['WIN_PCT'] * 100)



ab = AnnotationBbox(OffsetImage(plt.imread(f'/Users/franhuzjan/Downloads/wendy.png'), zoom=0.15), wendy_drafted, frameon=False)

ax.add_artist(ab)

offsetbox = TextArea(f"Drafted")

ab = AnnotationBbox(offsetbox, (wendy_drafted[0] - dx, wendy_drafted[1] - 0.02 * 100),
                box_alignment=(0., 0.5),
                arrowprops=dict(arrowstyle="->"), frameon=False)

ax.add_artist(ab)

champions = [1999, 2003, 2005, 2007, 2014]

for champion in champions:
    string_year = convert_int_to_season(champion)
    ax.scatter(champion, spurs[spurs['YEAR'] == string_year]['WIN_PCT'] * 100)
    
    if champion == 2003:
        ab = AnnotationBbox(OffsetImage(plt.imread(f'/Users/franhuzjan/Downloads/trophy.png'), zoom=0.15), (champion, spurs[spurs['YEAR'] == string_year]['WIN_PCT'] * 100 - 3), frameon=False)
    else:
        ab = AnnotationBbox(OffsetImage(plt.imread(f'/Users/franhuzjan/Downloads/trophy.png'), zoom=0.15), (champion, spurs[spurs['YEAR'] == string_year]['WIN_PCT'] * 100), frameon=False)

    ax.add_artist(ab)
    

ax.set_xlabel("Season", fontsize=25)
ax.set_ylabel("Win Percentage", fontsize=25)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.set_title("Can Wemdy bring back the glory?", fontdict={'fontsize': 35})
ab = AnnotationBbox(OffsetImage(plt.imread(f'twitter_logo (1).png'), zoom=0.1), (2022, 84), frameon=False)
ax.add_artist(ab)
ax.text(x=2026, y=83.75, ha='right', s='@FHuzjan', size=15)

plt.savefig('spurs_past_and_future.png', bbox_inches='tight')
```
